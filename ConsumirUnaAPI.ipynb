```python
# prompt: consumir una api con python

import requests

def consumir_api(url):
  """
  Consume una API dada su URL y retorna la respuesta.

  Args:
    url: La URL de la API a consumir.

  Returns:
    Un objeto de respuesta de la solicitud HTTP o None si ocurre un error.
  """
  try:
    response = requests.get(url)
    response.raise_for_status()  # Lanza una excepción para códigos de estado de error HTTP
    return response
  except requests.exceptions.RequestException as e:
    print(f"Error al consumir la API: {e}")
    return None

# Ejemplo de uso
#url_api = "https://jsonplaceholder.typicode.com/todos/1"  # Reemplaza con la URL de tu API

url_api = "https://newsapi.org/v2/everything?q=tesla&from=2024-09-23&sortBy=publishedAt&apiKey=1bb374c285834efc9859719fb8ab647e"

respuesta = consumir_api(url_api)

if respuesta:
  print("Código de estado:", respuesta.status_code)
  print("Contenido de la respuesta:", respuesta.json()) # Imprime la respuesta en formato JSON
  # Procesa la respuesta como necesites


payload={}
headers = {}

response = requests.request("GET", url_api, headers=headers, data=payload)

print(response.text)
```
